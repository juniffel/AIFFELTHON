```python
## 결측치 확인

for col in premier_league.columns:
  msg = 'column: {:>10}\t Percent of NaN value:{:.2f}%'.format(col, 100 * (premier_league[col].isnull().sum() / premier_league[col].shape[0]))
  print(msg)

msno.matrix(df = premier_league.iloc[:, :], figsize=(15, 15))
```

```python
# 컬럼 min이 결측치에 해당하는 경우의 인덱스 제거
a =  premier_league[premier_league["Min"].isnull()==True].index
print(a)

premier_league.drop(a, inplace=True)
premier_league.reset_index(drop=True, inplace=True)
premier_league.head()
```

```python
premier_league.to_csv('결측치처리.csv',index = False)
```
